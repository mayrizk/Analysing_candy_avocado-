```python
#now we need to figure out the distrubtion of species vs length and width 
# we will use the attrubite hue that takes a catogrical valrable that divide our distrubtion into categories by points colors 
sns.scatterplot(x='petal_length',y='petal_width',data=df,hue='species')
# we could see that it clrealy divide into cluster follow linner distruction 
#as setosa species"blue" clustering togather at 1:2 lenght vs 0.0:.6 width
# versicolor species"green" clustering togoter at 3:5.1 lenght vs 1.0:1.7 width
# viriginca species "red" clustring togother at 4.5 :6.9 lengh vs 1.4: 2.5 width 
# we  could came up with conclusion that the smallest petal is setosa (blue)
# the bigger petal is virgrinca (red)
# in between is the versicolor (green)
```
